## ABSTRACT

### Resumen

Este proyecto se centra en analizar la prevalencia y las causas de la diabetes en Estados Unidos, utilizando un conjunto de datos equilibrado basado en la encuesta BRFSS 2015 del CDC, disponible en Kaggle. La diabetes, una enfermedad crónica que impacta significativamente la salud pública y la economía, afecta a millones de personas en el país. En la primera fase, se busca analizar y visualizar hallazgos clave en los datos de más de 70,000 participantes, lo que permitirá en fases posteriores desarrollar modelos predictivos para mejorar el diagnóstico y tratamiento de la diabetes.

Dataset:
    Teboul, A. (2019). Diabetes Health Indicators Dataset. Kaggle. https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset


### Motivación:

 La diabetes es una enfermedad crónica que afecta a millones de personas en todo el mundo, con un impacto significativo en la calidad de vida de los pacientes y en los sistemas de salud pública. En los Estados Unidos, las tasas de diabetes y prediabetes están en aumento, lo que genera un costo económico y social alarmante. Este análisis está motivado por la necesidad de identificar los factores más influyentes en el desarrollo de la enfermedad y apoyar la creación de estrategias para su prevención y tratamiento, para posteriormente poder implementar dichas estrategias en difetentes países con el mismo problema.

 ### Audiencia: 
 
 Este proyecto está dirigido a profesionales de la salud, tomadores de decisiones en políticas públicas que buscan comprender mejor los factores de riesgo de la diabetes. Además, puede ser útil para estudiantes y académicos interesados en el análisis de datos y su aplicación en problemas de salud pública, así como para desarrolladores de soluciones tecnológicas orientadas a mejorar la prevención y detección de enfermedades crónicas.

### Hipótesis:
#### Basándonos en un análisis previo del conjunto de datos, se plantean las siguientes hipótesis:
- La prevalencia de diabetes varía según el nivel educativo.
- La edad y el género influyen en el padecimiento de la diabetes.
- Factores como la obesidad y la hipertensión aumentan el riesgo de diabetes.
- El consumo del alcohol y tabaco influyen en el padecimiento de la diabetes.


### Objetivos del proyecto:
- Probar las hipótesis mediante análisis estadísticos y visualizaciones.
- Identificar hallazgos importantes en los datos mediante visualizaciones.
- Mostrar la información de forma fácil y sencilla de comprender para la población  .

# Transformación y limpieza de los datos

```python
#Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importación de df

Path = "https://raw.githubusercontent.com/AdoCrz/Proyecto_DSI/main/Df_diabetes.csv"
df = pd.read_csv(Path)
```

```python
# Teniendo una vista rápida del df

df.head()
```

```python
# Información de la estructura del df
df.info()
```

```python
# Información de los Null
df.isnull().sum()
```

```python
# Creando las funciones para crear columnas categóricas que ayuden con la lectura de las gráficas y sea más fácil para los usuarios

def status(x):
    if x['Diabetes_binary'] == 0:
        return "No Diabetes"
    else:
        return "Diabetes"
# ----------------------------------------------------

def hip(x):
    if x['HighBP'] == 0:
        return "No Hipertenso"
    else:
        return "Hipertenso"
# ----------------------------------------------------

def col(x):
    if x['HighChol'] == 0:
        return "No Colesterol alto"
    else:
        return "Colesterol alto"
# ----------------------------------------------------

def Sm(x):
    if x['Smoker'] == 0:
        return "No Fumador"
    else:
        return "Fumador"
# ----------------------------------------------------

def Sx(x):
    if x['Sex'] == 0:
        return "Femenino"
    else:
        return "Masculino"
# ----------------------------------------------------

def af(x):
    if x['PhysActivity'] == 0:
        return "Actividad física inactiva"
    else:
        return "Actividad física activa"
# ----------------------------------------------------

def fr(x):
    if x['Fruits'] == 0:
        return "No consume frutas"
    else:
        return "Consume frutas"
# ----------------------------------------------------

def vg(x):
    if x['Veggies'] == 0:
        return "No consume vegetales"
    else:
        return "Consume vegetales"
# ----------------------------------------------------

def al(x):
    if x['AnyHealthcare'] == 0:
        return "No tiene atención médica"
    else:
        return "Tiene atención médica"
# ----------------------------------------------------

def education(x):
    if x['Education'] == 1:
        return "No se graduó de la escuela secundaria"
    elif x['Education'] == 2:
        return "Graduado de la escuela secundaria"
    elif x['Education'] == 3:
        return "Asistió a la universidad"
    elif x['Education'] == 4:
        return "Graduado universitario"
    elif x['Education'] == 5:
        return "Estudios de posgrado"
    elif x['Education'] == 6:
        return "Doctorado o equivalente"
# ----------------------------------------------------

def age(x):
    if x['Age'] == 1:
        return "18-24 años"
    elif x['Age'] == 2:
        return "25-29 años"
    elif x['Age'] == 3:
        return "30-34 años"
    elif x['Age'] == 4:
        return "35-39 años"
    elif x['Age'] == 5:
        return "40-44 años"
    elif x['Age'] == 6:
        return "45-49 años"
    elif x['Age'] == 7:
        return "50-54 años"
    elif x['Age'] == 8:
        return "55-59 años"
    elif x['Age'] == 9:
        return "60-64 años"
    elif x['Age'] == 10:
        return "65-69 años"
    elif x['Age'] == 11:
        return "70-74 años"
    elif x['Age'] == 12:
        return "75-79 años"
    elif x['Age'] == 13:
        return "80 o más años"
```

```python
# Creando las variables categóricas

df['Status'] = df.apply(status, axis = 1)

df['Hipertenso'] = df.apply(hip, axis = 1)

df['Nivel_Col'] = df.apply(col, axis = 1)

df['Género'] = df.apply(Sx, axis = 1)

df['Actividad_física'] = df.apply(af, axis = 1)

df['Consumo_Frutas'] = df.apply(af, axis = 1)

df['Consumo_vegetales'] = df.apply(fr, axis = 1)

df['Seguro_med'] = df.apply(al, axis = 1)

df['Educación'] = df.apply(education, axis = 1)

df['Edad'] = df.apply(age, axis = 1)
```

### Interpretación de Resultados

1. **BMI (Índice de Masa Corporal):**

2. **GenHlth (Salud General):**
   - **Media:** 2.84 - Salud general promedio entre "buena" y "regular", en una escala de 1 (excelente) a 5 (pobre).
   - **Distribución:** Mayoría con percepción de salud entre "buena" y "regular".

3. **MentHlth (Salud Mental - días de mala salud mental):**
   - **Mediana:** 0 - La mayoría no reporta días de mala salud mental, aunque hay casos severos con hasta 30 días.

4. **PhysHlth (Salud Física - días de mala salud física):**
   - **Mediana:** 0 - Similar a la salud mental, la mayoría no reporta problemas, pero algunos tienen problemas físicos graves.

5. **Age (Edad):**
   - **Media:** 8.58 - Promedio en la categoría de 55-59 años, indicando una muestra mayor.
   - **Distribución:** Bien distribuida, pero sesgada hacia una población mayor.

6. **Education (Educación):**
   - **Media:** 4.92 - La mayoría tiene educación superior, con muchos completando la universidad o estudios de posgrado.

7. **Income (Ingresos):**
   - **Media:** 5.70 - Ingreso promedio entre $35,000 y $50,000 anuales, indicando una clase media predominante.

La muestra tiene un sobrepeso significativo, está orientada hacia una población mayor con buena educación y un ingreso de clase media. Hay poca prevalencia de problemas de salud mental y física reportados, aunque hay casos extremos en ambos.

## Análisis gráfico (univariado y multivariado)

```python
# Gráficas para analizar la primer y segunda hipótesis

# Configuración de los gráficos
fig, ax = plt.subplots(figsize=(10, 6))

# Prevalencia de diabetes según la edad
age_diabetes = df.groupby('Edad')['Diabetes_binary'].mean()
ax.bar(age_diabetes.index, age_diabetes.values, color='#265C4B')
ax.set_title('Prevalencia de Diabetes según la Edad')
ax.set_xlabel('Edad')
ax.set_ylabel('Proporción de Diabetes')
ax.tick_params(axis='x', rotation=45)

# Mostrar el gráfico
plt.show()
```

```python

# Configuración de los gráficos

fig, ax = plt.subplots(figsize=(10, 5))

# Distribución de personas con diabetes por nivel educativo
df_no_diabetes = df[df['Diabetes_binary'] == 0]['Education']
df_diabetes = df[df['Diabetes_binary'] == 1]['Education']

ax.hist(
    [df_no_diabetes, df_diabetes],
    bins=6,
    stacked=True,
    color=['#265C4B', '#9FC131'],
    label=['No Diabetes', 'Diabetes'],
    edgecolor='black'
)
ax.set_title('Distribución del Nivel Educativo')
ax.set_xlabel('Nivel Educativo')
ax.set_ylabel('Frecuencia')
ax.legend()

# Mostrar el gráfico
plt.show()
```

```python
# Configuración del gráfico único
fig, ax = plt.subplots(figsize=(10, 6))

# Prevalencia de diabetes según la educación
education_diabetes = df.groupby('Educación')['Diabetes_binary'].mean()
ax.bar(education_diabetes.index, education_diabetes.values, color='#9FC131')
ax.set_title('Prevalencia de Diabetes según la Educación')
ax.set_xlabel('Nivel Educativo')
ax.set_ylabel('Proporción de Diabetes')
ax.tick_params(axis='x', rotation=45)

# Etiquetas de las barras
for i, v in enumerate(education_diabetes.values):
    ax.text(i, v + 0.01, f"{v:.2f}", ha='center')

# Mostrar el gráfico
plt.show()
```

### Interpretación Relacionada con las Primeras Hipótesis

**Prevalencia de Diabetes según la Edad**

La prevalencia de diabetes aumenta significativamente con la edad, confirmando la hipótesis de que la edad influye en la probabilidad de desarrollar diabetes. La proporción pasa de un 8-10% en edades jóvenes (18-29 años) a más del 60% en los grupos mayores (65-79 años).

**Distribución del Nivel Educativo**

Aunque la mayoría de los participantes con un nivel educativo más alto (5 y 6) no tienen diabetes, hay una cantidad considerable en estos grupos que sí la tienen, lo que sugiere una posible influencia del nivel educativo en la prevalencia de diabetes.

**Prevalencia de Diabetes según la Educación**

La educación muestra un patrón interesante; pues la mayor prevalencia se observa en quienes asistieron a la universidad pero no se graduaron (67%) y en graduados de la escuela secundaria (72%), lo que podría indicar que ciertos niveles educativos están más relacionados con un mayor riesgo de diabetes.
